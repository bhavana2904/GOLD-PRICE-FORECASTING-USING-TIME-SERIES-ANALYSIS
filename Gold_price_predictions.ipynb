```python
# Make sure Date is the index
df.set_index('Date', inplace=True)

# Extract just the Price column as a series
price_series = df['Price']

print("Data prepared for ARIMA:")
print(price_series.head())
print(f"\nTotal observations: {len(price_series)}")
```

```python
# Test if data is stationary using ADF test
def test_stationarity(series):
    result = adfuller(series)

    print('ADF Statistic:', result[0])
    print('p-value:', result[1])

# Test original price series
print("Testing Original Price Data:")
print("="*50)
is_stationary = test_stationarity(price_series)
```

as p-value>0.5 that means stationary is present in data

now we will make the data stationary using differencing

```python
train_diff = train_data.diff().dropna()

print("Testing Original Price Data:")
print("="*50)
is_stationary = test_stationarity(price_series)

print("Original vs Differenced Data:")
print("="*50)
print("\nOriginal prices (first 5):")
print(price_series.head())
print("\nDifferenced prices (first 5) - Daily changes:")
print(train_diff.head())

# Now test if differenced data is stationary
print("\n" + "="*50)
print("Testing Differenced Data:")
print("="*50)
is_stationary_now = test_stationarity(train_diff)
```

```python
# Plot ACF and PACF for TRAINING data
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# ACF plot - helps find 'q'
plot_acf(train_diff, lags=30, ax=axes[0])
axes[0].set_title('ACF Plot - Find q (Training Data)', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Lags')

# PACF plot - helps find 'p'
plot_pacf(train_diff, lags=30, ax=axes[1])
axes[1].set_title('PACF Plot - Find p (Training Data)', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Lags')

plt.tight_layout()
plt.show()
```

p=1, q=1 and d=1

```python
model = ARIMA(train_data, order=(1, 1, 1))
model_fit = model.fit()
print(model_fit.summary())
```

```python
# Predict for the length of test data
predictions = model_fit.forecast(steps=len(test_data))

# Calculate errors
mse = mean_squared_error(test_data, predictions)
rmse = np.sqrt(mse)
mae = mean_absolute_error(test_data, predictions)
mape = np.mean(np.abs((test_data - predictions) / test_data)) * 100

print(f"\n Model Performance:")
print(f"  RMSE (Root Mean Squared Error): ${rmse:.2f}")
print(f"  MAE (Mean Absolute Error): ${mae:.2f}")
print(f"  MAPE (Mean Absolute % Error): {mape:.2f}%")

# Visualize predictions
plt.figure(figsize=(16, 6))
plt.plot(train_data.index, train_data.values, label='Training Data', color='blue', linewidth=2)
plt.plot(test_data.index, test_data.values, label='Actual Test Data', color='green', linewidth=2)
plt.plot(test_data.index, predictions, label='ARIMA Predictions', color='red', linewidth=2, linestyle='--')
plt.axvline(x=train_data.index[-1], color='black', linestyle=':', linewidth=2, label='Train/Test Split')
plt.title('ARIMA(1,1,1) Gold Price Prediction', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Gold Price ($)', fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Try different ARIMA parameters
param_combinations = [
    (1,1,1), (1,1,2), (2,1,1), (2,1,2),
    (1,1,0), (0,1,1), (3,1,1), (1,1,3)
]

results = []

print("Testing Multiple ARIMA Models...")
print("="*70)

for params in param_combinations:
    try:
        model = ARIMA(train_data, order=params)
        model_fit = model.fit()
        predictions = model_fit.forecast(steps=len(test_data))

        mse = mean_squared_error(test_data, predictions)
        rmse = np.sqrt(mse)
        mae = mean_absolute_error(test_data, predictions)

        results.append({
            'Model': f'ARIMA{params}',
            'RMSE': rmse,
            'MAE': mae
        })
        print(f" ARIMA{params} - RMSE: ${rmse:.2f}, MAE: ${mae:.2f}")
    except:
        print(f" ARIMA{params} - Failed to converge")

# Show best models
results_df = pd.DataFrame(results).sort_values('RMSE')
print("\n" + "="*70)
print(" Best Models (Ranked by RMSE):")
print("="*70)
print(results_df.to_string(index=False))
print(f"\n Best Model: {results_df.iloc[0]['Model']} with RMSE ${results_df.iloc[0]['RMSE']:.2f}")
```

```python
# Build best model ARIMA(1,1,0)
print("Building Best Model: ARIMA(1,1,0)")
print("="*70)

best_model = ARIMA(train_data, order=(1,1,0))
best_model_fit = best_model.fit()

# Predictions on test data
test_predictions = best_model_fit.forecast(steps=len(test_data))

# Calculate final metrics
rmse = np.sqrt(mean_squared_error(test_data, test_predictions))
mae = mean_absolute_error(test_data, test_predictions)

print(f"\n Final Model Performance:")
print(f"  RMSE: ${rmse:.2f}")
print(f"  MAE: ${mae:.2f}")

# Now predict FUTURE (next 30 days)
print("\n" + "="*70)
print(" PREDICTING NEXT 30 DAYS")
print("="*70)

# Retrain on ALL data for future predictions
final_model = ARIMA(price_series, order=(1,1,0))
final_model_fit = final_model.fit()

# Forecast next 30 days
future_steps = 30
future_predictions = final_model_fit.forecast(steps=future_steps)
future_dates = pd.date_range(start=price_series.index[-1] + pd.Timedelta(days=1), periods=future_steps)

# Create future dataframe
future_df = pd.DataFrame({
    'Date': future_dates,
    'Predicted_Price': future_predictions
})

print("\nNext 30 Days Prediction:")
print(future_df.head(10))

# Visualization
fig, axes = plt.subplots(2, 1, figsize=(16, 10))

# Plot 1: Historical + Test Predictions
axes[0].plot(train_data.index, train_data.values, label='Training Data', color='blue', linewidth=2)
axes[0].plot(test_data.index, test_data.values, label='Actual Test Data', color='green', linewidth=2)
axes[0].plot(test_data.index, test_predictions, label='Test Predictions', color='red', linewidth=2, linestyle='--')
axes[0].axvline(x=train_data.index[-1], color='black', linestyle=':', linewidth=2, label='Train/Test Split')
axes[0].set_title('ARIMA(1,1,0) - Historical Prediction Performance', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Date')
axes[0].set_ylabel('Gold Price ($)')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Plot 2: Recent Data + Future Forecast
recent_data = price_series[-90:]  # Last 90 days
axes[1].plot(recent_data.index, recent_data.values, label='Recent Historical Data', color='blue', linewidth=2)
axes[1].plot(future_dates, future_predictions, label='30-Day Forecast', color='red', linewidth=2, marker='o', markersize=3)
axes[1].axvline(x=price_series.index[-1], color='black', linestyle=':', linewidth=2, label='Today')
axes[1].set_title('ARIMA(1,1,0) - Next 30 Days Forecast', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Gold Price ($)')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Summary
last_price = price_series.iloc[-1]
predicted_price_30d = future_predictions.iloc[-1]
change = predicted_price_30d - last_price
change_pct = (change / last_price) * 100

print("\n" + "="*70)
print(" FORECAST SUMMARY")
print("="*70)
print(f"Current Gold Price: ${last_price:.2f}")
print(f"Predicted Price (30 days): ${predicted_price_30d:.2f}")
print(f"Expected Change: ${change:+.2f} ({change_pct:+.2f}%)")

if change > 0:
    print(" Trend: UPWARD - Gold prices expected to RISE")
else:
    print(" Trend: DOWNWARD - Gold prices expected to FALL")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet
from sklearn.metrics import mean_squared_error, mean_absolute_error
from statsmodels.tsa.arima.model import ARIMA
import warnings
warnings.filterwarnings('ignore')

print("="*70)
print("PROPHET vs ARIMA: GOLD PRICE PREDICTION COMPARISON")
print("="*70)

# Prepare data for Prophet (needs 'ds' and 'y' columns)
df_prophet = pd.DataFrame({
    'ds': price_series.index,
    'y': price_series.values
})

print(f"\nDataset: {len(df_prophet)} observations")
print(f"Date range: {df_prophet['ds'].min()} to {df_prophet['ds'].max()}")

# Split data (same 80/20 split as ARIMA)
train_size = int(len(df_prophet) * 0.8)
train_prophet = df_prophet[:train_size]
test_prophet = df_prophet[train_size:]

print(f"\nTraining: {len(train_prophet)} | Testing: {len(test_prophet)}")

# ============================================================================
# BUILD PROPHET MODEL
# ============================================================================
print("\n" + "="*70)
print("TRAINING PROPHET MODEL")
print("="*70)

# Initialize and train Prophet
prophet_model = Prophet(
    daily_seasonality=False,
    weekly_seasonality=True,
    yearly_seasonality=True,
    changepoint_prior_scale=0.05  # Controls trend flexibility
)

prophet_model.fit(train_prophet)
print(" Prophet model trained successfully!")

# Make predictions on test period
future_test = test_prophet[['ds']].copy()
prophet_predictions = prophet_model.predict(future_test)

# Extract predicted values
prophet_pred_values = prophet_predictions['yhat'].values
prophet_actual_values = test_prophet['y'].values

# Calculate Prophet metrics
prophet_rmse = np.sqrt(mean_squared_error(prophet_actual_values, prophet_pred_values))
prophet_mae = mean_absolute_error(prophet_actual_values, prophet_pred_values)
prophet_mape = np.mean(np.abs((prophet_actual_values - prophet_pred_values) / prophet_actual_values)) * 100

print(f"\n Prophet Performance on Test Data:")
print(f"  RMSE: ${prophet_rmse:.2f}")
print(f"  MAE: ${prophet_mae:.2f}")
print(f"  MAPE: {prophet_mape:.2f}%")

# ============================================================================
# BUILD ARIMA MODEL (for comparison)
# ============================================================================
print("\n" + "="*70)
print("TRAINING ARIMA(1,1,0) MODEL")
print("="*70)

arima_model = ARIMA(train_data, order=(1,1,0))
arima_model_fit = arima_model.fit()
arima_predictions = arima_model_fit.forecast(steps=len(test_data))

# Calculate ARIMA metrics
arima_rmse = np.sqrt(mean_squared_error(test_data.values, arima_predictions))
arima_mae = mean_absolute_error(test_data.values, arima_predictions)
arima_mape = np.mean(np.abs((test_data.values - arima_predictions) / test_data.values)) * 100

print(f"\n ARIMA Performance on Test Data:")
print(f"  RMSE: ${arima_rmse:.2f}")
print(f"  MAE: ${arima_mae:.2f}")
print(f"  MAPE: {arima_mape:.2f}%")

# ============================================================================
# MODEL COMPARISON
# ============================================================================
print("\n" + "="*70)
print(" MODEL COMPARISON")
print("="*70)

comparison_df = pd.DataFrame({
    'Model': ['Prophet', 'ARIMA(1,1,0)'],
    'RMSE': [prophet_rmse, arima_rmse],
    'MAE': [prophet_mae, arima_mae],
    'MAPE (%)': [prophet_mape, arima_mape]
})

print("\n", comparison_df.to_string(index=False))

# Determine winner
if prophet_rmse < arima_rmse:
    print(f"\n Winner: PROPHET (RMSE ${prophet_rmse:.2f} vs ${arima_rmse:.2f})")
    improvement = ((arima_rmse - prophet_rmse) / arima_rmse) * 100
    print(f"   Prophet is {improvement:.1f}% more accurate!")
    best_model = 'Prophet'
else:
    print(f"\n Winner: ARIMA (RMSE ${arima_rmse:.2f} vs ${prophet_rmse:.2f})")
    improvement = ((prophet_rmse - arima_rmse) / prophet_rmse) * 100
    print(f"   ARIMA is {improvement:.1f}% more accurate!")
    best_model = 'ARIMA'

# ============================================================================
# VISUALIZATIONS
# ============================================================================
fig = plt.figure(figsize=(16, 12))
gs = fig.add_gridspec(3, 2, hspace=0.3, wspace=0.3)

# Plot 1: Prophet Predictions
ax1 = fig.add_subplot(gs[0, :])
ax1.plot(train_prophet['ds'], train_prophet['y'], label='Training Data', color='blue', linewidth=2)
ax1.plot(test_prophet['ds'], test_prophet['y'], label='Actual Test Data', color='green', linewidth=2)
ax1.plot(test_prophet['ds'], prophet_pred_values, label='Prophet Predictions',
         color='red', linewidth=2, linestyle='--', alpha=0.8)
ax1.axvline(x=train_prophet['ds'].iloc[-1], color='black', linestyle=':', linewidth=2, label='Train/Test Split')
ax1.set_title('Prophet Model - Predictions vs Actual', fontsize=14, fontweight='bold')
ax1.set_xlabel('Date')
ax1.set_ylabel('Gold Price ($)')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot 2: ARIMA Predictions
ax2 = fig.add_subplot(gs[1, :])
ax2.plot(train_data.index, train_data.values, label='Training Data', color='blue', linewidth=2)
ax2.plot(test_data.index, test_data.values, label='Actual Test Data', color='green', linewidth=2)
ax2.plot(test_data.index, arima_predictions, label='ARIMA Predictions',
         color='orange', linewidth=2, linestyle='--', alpha=0.8)
ax2.axvline(x=train_data.index[-1], color='black', linestyle=':', linewidth=2, label='Train/Test Split')
ax2.set_title('ARIMA(1,1,0) Model - Predictions vs Actual', fontsize=14, fontweight='bold')
ax2.set_xlabel('Date')
ax2.set_ylabel('Gold Price ($)')
ax2.legend()
ax2.grid(True, alpha=0.3)

# Plot 3: Model Comparison - RMSE
ax3 = fig.add_subplot(gs[2, 0])
models = ['Prophet', 'ARIMA']
rmse_values = [prophet_rmse, arima_rmse]
colors = ['#2ecc71' if prophet_rmse < arima_rmse else '#e74c3c',
          '#e74c3c' if prophet_rmse < arima_rmse else '#2ecc71']
bars = ax3.bar(models, rmse_values, color=colors, alpha=0.7, edgecolor='black', linewidth=2)
ax3.set_title('RMSE Comparison (Lower is Better)', fontsize=12, fontweight='bold')
ax3.set_ylabel('RMSE ($)')
for i, (bar, val) in enumerate(zip(bars, rmse_values)):
    ax3.text(bar.get_x() + bar.get_width()/2, val + 50, f'${val:.2f}',
             ha='center', va='bottom', fontweight='bold', fontsize=11)
ax3.grid(True, alpha=0.3, axis='y')

# Plot 4: Prediction Error Distribution
ax4 = fig.add_subplot(gs[2, 1])
prophet_errors = prophet_actual_values - prophet_pred_values
arima_errors = test_data.values - arima_predictions
ax4.hist(prophet_errors, bins=30, alpha=0.6, label='Prophet Errors', color='green', edgecolor='black')
ax4.hist(arima_errors, bins=30, alpha=0.6, label='ARIMA Errors', color='orange', edgecolor='black')
ax4.axvline(x=0, color='red', linestyle='--', linewidth=2)
ax4.set_title('Prediction Error Distribution', fontsize=12, fontweight='bold')
ax4.set_xlabel('Prediction Error ($)')
ax4.set_ylabel('Frequency')
ax4.legend()
ax4.grid(True, alpha=0.3)

plt.suptitle('Gold Price Prediction: Prophet vs ARIMA Comparison',
             fontsize=16, fontweight='bold', y=0.995)
plt.show()

# ============================================================================
# FUTURE FORECASTING (30 DAYS) - Best Model
# ============================================================================
print("\n" + "="*70)
print(f" FUTURE FORECAST (30 DAYS) - Using {best_model}")
print("="*70)

if best_model == 'Prophet':
    # Retrain Prophet on full data
    final_prophet = Prophet(daily_seasonality=False, weekly_seasonality=True, yearly_seasonality=True)
    final_prophet.fit(df_prophet)

    # Create future dates
    future = final_prophet.make_future_dataframe(periods=30)
    forecast = final_prophet.predict(future)

    # Extract last 30 days forecast
    future_forecast = forecast.tail(30)[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
    future_forecast.columns = ['Date', 'Predicted_Price', 'Lower_Bound', 'Upper_Bound']

    print("\nNext 30 Days Forecast (with confidence intervals):")
    print(future_forecast.head(10).to_string(index=False))

    # Plot future forecast
    fig, ax = plt.subplots(figsize=(16, 6))

    # Recent data
    recent_data = df_prophet.tail(90)
    ax.plot(recent_data['ds'], recent_data['y'], label='Recent Historical Data',
            color='blue', linewidth=2)

    # Future forecast
    ax.plot(future_forecast['Date'], future_forecast['Predicted_Price'],
            label='30-Day Forecast', color='red', linewidth=2, marker='o', markersize=4)
    ax.fill_between(future_forecast['Date'],
                     future_forecast['Lower_Bound'],
                     future_forecast['Upper_Bound'],
                     alpha=0.3, color='red', label='Confidence Interval')

    ax.axvline(x=df_prophet['ds'].iloc[-1], color='black', linestyle=':', linewidth=2, label='Today')
    ax.set_title('Prophet - Next 30 Days Gold Price Forecast', fontsize=14, fontweight='bold')
    ax.set_xlabel('Date')
    ax.set_ylabel('Gold Price ($)')
    ax.legend()
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

    # Summary
    last_price = df_prophet['y'].iloc[-1]
    predicted_30d = future_forecast['Predicted_Price'].iloc[-1]
    change = predicted_30d - last_price
    change_pct = (change / last_price) * 100

else:  # ARIMA
    final_arima = ARIMA(price_series, order=(1,1,0))
    final_arima_fit = final_arima.fit()
    future_arima = final_arima_fit.forecast(steps=30)

    last_price = price_series.iloc[-1]
    predicted_30d = future_arima.iloc[-1]
    change = predicted_30d - last_price
    change_pct = (change / last_price) * 100

print("\n" + "="*70)
print(" FORECAST SUMMARY")
print("="*70)
print(f"Current Gold Price: ${last_price:.2f}")
print(f"Predicted Price (30 days): ${predicted_30d:.2f}")
print(f"Expected Change: ${change:+.2f} ({change_pct:+.2f}%)")

if change > 0:
    print(" Trend: UPWARD - Gold prices expected to RISE")
else:
    print(" Trend: DOWNWARD - Gold prices expected to FALL")

print("\n" + "="*70)
print(" ANALYSIS COMPLETE - Both Models Compared!")
print("="*70)
```
